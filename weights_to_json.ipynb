```python
import os
import json
from ultralytics import YOLO

# Load a pretrained YOLO model
model = YOLO('C:/lcy/2023_final/vehicle_detection/runs/detect/train6/weights/epoch19.pt')
sequence_name = 'yolov8l_bonus_train3_best'

image_directory = 'C:/lcy/2023_final/Bonus_Image'
output_file_path = sequence_name + "_pred.json"

predictions = []

# Get all image in the image_directory_folder

img_path = sorted(os.listdir(image_directory))

# Loop through each file in the directory

for filename in img_path:

    if filename.endswith(".png"): 

        file_path = os.path.join(image_directory, filename)
        sample_token = filename.split('.')[0]

        results = model(file_path)
        

        for i in range(len(results[0].boxes.xyxy)):

            x_min, y_min, x_max, y_max = results[0].boxes.xyxy[i].cpu()
            x_min, y_min, x_max, y_max = float(x_min), float(y_min), float(x_max), float(y_max) 
            score = float(results[0].boxes.conf[i].cpu())
            coor = [[x_min, y_min],
                    [x_min, y_max],
                    [x_max, y_max],
                    [x_max, y_min]]
            predicts_object = {'sample_token': f'{sample_token}', 'points': coor, 'name': "car", 'score': score}
            predictions.append(predicts_object)                
        

# Write to JSON only once after all files have been processed
with open(output_file_path, "w") as outfile:
    json.dump(predictions, outfile, indent=2)
```
